```python
# Importando o Folium

from folium import Choropleth, Marker
from folium.plugins import MarkerCluster
```

```python
import folium
center = [-23.60, -46.6]

# Create a Open Street Map
m_open_street = folium.Map(location=center, 
                 width=500,height=500,
                 tiles='OpenStreetMap', zoom_start=7)

# Display the map

m_open_street
```

```python
dados_geo['centroid'] = dados_geo['Point'].apply(lambda z: (z.centroid.y, z.centroid.x))
```

```python
for _, imovel in dados_geo.iterrows():
    folium.Marker(
        location=[imovel['latitude'], imovel['longitude']],
    ).add_to(m_open_street)

m_open_street
```

```python
# Add points to the map
mc = MarkerCluster()
for index, row in dados_geo.iterrows():
    mc.add_child(Marker(row['centroid'], popup=row['Bairro']))

m_open_street.add_child(mc)

# Display the map
m_open_street
```

```python
# Create the map
m_stamen_2 = folium.Map(location=center, 
                 tiles='Stamen Toner', zoom_start=8)

m_stamen_2
```
